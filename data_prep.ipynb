```python
import pandas as pd
import numpy as np
```

```python
print("Loading Hospital dataset ...")
df = pd.read_csv(r"C:\Users\Admin\Desktop\xpecto\Hospital ER.csv")
print(f"  Raw shape: {df.shape[0]} rows × {df.shape[1]} columns")
 
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d %H:%M:%S")
print("  Date column standardised to ISO format")
```

```python
df.isna().sum()
```

```python
median_sat = df["Satisfaction Score"].median()
df["Satisfaction Score"] = df["Satisfaction Score"].fillna(median_sat)
print(f" Satisfaction Score nulls filled with median ({median_sat})")
```

```python
df["Gender"] = df["Gender"].replace("NC", "Unknown")
print("  Gender 'NC' changed to 'Unknown'")
```

```python
df.drop(columns=["Patient ID", "First Name", "Last Name"], inplace=True)
print("  Dropped PII columns (Patient ID, First Name, Last Name)") #they add add zero ML signal
```

```python
df.columns = [col.replace(" ", "_") for col in df.columns]
print("Column name spaces replaced with underscores")
```

```python
output = "hospital_er_clean.csv"
df.to_csv(output, index=False)
```
